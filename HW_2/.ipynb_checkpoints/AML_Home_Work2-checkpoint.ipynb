## Step - 1: Read data into pandas dataframe

```python
# Download data	from https://ndownloader.figshare.com/files/7586326
import pandas as pd 

# read data into a pandas data-frame
mydata=pd.read_csv("https://ndownloader.figshare.com/files/7586326")

```

# Step - 2: Data Understanding

```python
# Basic Statistics of Columns and understanding of dataset
mydata.head(n=10) 
mydata.shape # Rows=15,342 Columns = 197

# data types of variables
mydata.dtypes

# Not applicable count in rent variable
print len(mydata[mydata['uf28']==9999])
# 5113 rows of data not applicable

# Identify Variables with low variance
summary=pd.DataFrame(mydata.apply(lambda x: 100*x.value_counts(normalize=True)).T.stack(),columns=['fill-rate'])
summary[summary['fill-rate']>95]
```
